## Staking Transactions & Staking Balance

### Transaction List - Total Amount Staked by P2P Delegators
https://docs.berascan.com/api-endpoints/accounts#get-a-list-of-normal-transactions-by-address

```python
pd.set_option('display.float_format', '{:.1f}'.format)

# parameters for the API request
params = {
    'module': 'account',
    'action': 'txlist',
    'startblock': 0,
    'endblock': 99999999,
    'page': 1,
    'offset': 10000,  # Number of transactions to retrieve per page
    'sort': 'asc',  # Sort order: asc or desc
    'apikey': api_key 
}

all_staking_transactions = []
for address in val_del_addresses_lower: #[:-1] excludes the p2p validator address in the list.
    params['address'] = address 
    
    response = requests.get(api_url, params=params)
    
    if response.status_code == 200:
        data = response.json()
        
        if 'result' in data:
            transactions = data['result']
            for txn in transactions:
                    txn['delegator_address'] = address  
                    txn['ChainName'] = 'Berachain'
                    all_staking_transactions.append(txn)
        else:
            print(f"No transactions found for address {address}")
    else:
        print(f"Error: {response.status_code} for address {address}")

address_label= {
    val_del_addresses_lower[1] : 'delegator1',
    val_del_addresses_lower[2] : 'delegator2',
    val_del_addresses_lower[3] : 'delegator3',
    val_del_addresses_lower[4] : 'delegator4',
    val_del_addresses_lower[5] : 'delegator5',
    val_del_addresses_lower[6] : 'delegator6',
    val_del_addresses_lower[0] : 'Operator Validator'
}
df_txlist = pd.DataFrame(all_staking_transactions)
df_txlist['value_bera'] = df_txlist['value'].astype(float) / 1e18 #converting value to Bera.
df_txlist['value_bera'] = df_txlist['value_bera'].round(6)
df_txlist['timeStamp'] = pd.to_datetime(df_txlist['timeStamp'], unit='s')
df_txlist = df_txlist.sort_values(by='timeStamp')
df_txlist = df_txlist[df_txlist['to'] == Berachain_Contract_Addresses[0]]
df_txlist['Client'] = df_txlist['from'].map(address_label)
df_txlist['Staked_Bal_History'] = df_txlist['value_bera'].cumsum()

# df_txlist.to_csv('mac_local_path/berachain_delegators_staking.csv', index=False)

df_txlist
```

### BGT Rewards & Incentive Token Rewards

- https://docs.berascan.com/api-endpoints/tokens#get-erc20-token-account-balance-for-tokencontractaddress 
- https://docs.berascan.com/api-endpoints/accounts#get-a-list-of-erc20-token-transfer-events-by-address 

Since we are capturing token rewards transferred to our validator from about 13 token contract addresses,
we could be looking at well over 1,500,000 records of historic transactions from this api call in the script below.

The approach here taken was to create a for loop and block numbers to around api limitations

```python
import requests
import pandas as pd
import time

OFFSET = 10000
MAX_PAGES = 1  # keep OFFSET * MAX_PAGES ≤ 10000
START_BLOCK = 5769612
TARGET_END_BLOCK = 7112660

ERC20TokenTransferEvents = []

for contract_address in Berachain_Contract_Addresses:
    current_start_block = START_BLOCK
    done = False

    while not done and current_start_block <= TARGET_END_BLOCK:
        total_fetched = 0
        latest_block_in_batch = current_start_block

        for page in range(1, MAX_PAGES + 1):
            params_txn = {
                'module': 'account',
                'action': 'tokentx',
                'contractaddress': contract_address,
                'address': val_del_addresses_lower[0],
                'page': page,
                'offset': OFFSET,
                'startblock': current_start_block,
                'endblock': TARGET_END_BLOCK,
                'sort': 'asc',
                'apikey': api_key
            }

            response_txn = requests.get(api_url, params=params_txn)

            if response_txn.status_code == 200:
                data_txn = response_txn.json()
                
                result = data_txn.get("result", "")

                # Check if result is a list of transactions
                if isinstance(result, list) and result:
                    transactions = result
                    total_fetched += len(transactions)

                    for txn in transactions:
                        txn['validator_address'] = val_del_addresses_lower[0]
                        block_number = int(txn['blockNumber'])
                        latest_block_in_batch = max(latest_block_in_batch, block_number)
                        ERC20TokenTransferEvents.append(txn)

                    print(f"Fetched page {page} for contract {contract_address}, from block {current_start_block}")
                    time.sleep(0.2)

                else:
                    print(f"No more data or unexpected result for {contract_address} at page {page}, block {current_start_block}")
                    print("Raw result:", result)
                    done = True
                    break

            else:
                print(f"Error {response_txn.status_code} for contract {contract_address}")
                done = True
                break

        if total_fetched >= OFFSET * MAX_PAGES:
            if latest_block_in_batch >= TARGET_END_BLOCK:
                print(f"Reached or passed END_BLOCK {TARGET_END_BLOCK} for {contract_address}.")
                done = True
            else:
                current_start_block = latest_block_in_batch + 1
                print(f"Moving to next batch starting from block {current_start_block} for {contract_address}")
        else:
            print(f"Fewer than {OFFSET * MAX_PAGES} records fetched. Ending fetch for {contract_address}.")
            done = True

# Convert to DataFrame
token_rewards_batched = pd.DataFrame(ERC20TokenTransferEvents)

if not token_rewards_batched.empty:
    token_rewards_batched['blockNumber'] = token_rewards_batched['blockNumber'].astype(int)
    token_rewards_batched['timeStamp'] = pd.to_datetime(token_rewards_batched['timeStamp'], unit='s')
    token_rewards_batched = token_rewards_batched.sort_values(by='timeStamp', ascending=False)

    # Optional: save to CSV
    # token_rewards_batched.to_csv('mac_local_path/berachain_ALLToken_rewardsJUNEb.csv', index=False)

else:
    print("No transactions fetched. DataFrame is empty.")
```
